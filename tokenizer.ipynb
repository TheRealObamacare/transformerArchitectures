```python
with open('input.txt') as file:
    text = file.read(24597)    
tokens = text.encode('utf-8')
tokens = list(map(int, tokens))
print(tokens)
print("\nLength of tokens:", len(tokens))
```

```python
def getPairs(text):
    pairs = {}
    for pair in zip(text, text[1:]):
        pairs[pair] = pairs.get(pair, 0) + 1
    return pairs

pairs = getPairs(tokens)
pairs
```

```python
mostCommon = max(pairs, key=pairs.get)
mostCommon
```

```python
def merge(tokens, pair, idx):
    newids = []
    i = 0
    while i < len(tokens):
        if i < len(tokens) - 1 and (tokens[i], tokens[i+1]) == pair:
            newids.append(idx)
            i += 2
        else:
            newids.append(tokens[i])
            i += 1
    
    return newids

vocabSize = 276
numMerges = vocabSize - 256
merges = {}
ids = list(tokens)
for i in range(numMerges):
    pairs = getPairs(ids)
    bestPair = max(pairs, key=pairs.get)
    idx = 256 + i
    print(f"Merging {bestPair} into a new token at index {idx}")
    ids = merge(ids, bestPair, idx)
    merges[bestPair] = idx
```

```python
print(f"tokens length: {len(tokens)} \nids length: {len(ids)}\ncompression ratio: {len(tokens) / len(ids):.2f}X")
```

```python
vocabList = {idx: bytes([idx]) for idx in range(256)}
for (ch1,ch2) ,idx in merges.items():
    vocabList[idx] = bytes(ch1) + bytes(ch2)
def decode(ids):
    text = b"".join(vocabList[idx] for idx in ids)
    return text.decode("UTF-8", errors = "replace")

decode(tokens)
```

```python
def encode(text):
    tokens = list(text.encode("UTF-8"))
    for i in range(10):
        pairs = getPairs(tokens)
        bestPair = max(pairs, key=pairs.get)
        idx = 256 + i
        encodedText = merge(tokens, bestPair, idx)
    return encodedText

print(encode("hello world"))
```
